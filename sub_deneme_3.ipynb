```python
def utc_to_unix_millis_array(utc_millis_array):
    # Unix epoch start time in milliseconds
    unix_epoch_start = np.datetime64('1970-01-01T00:00:00', 'ms').astype('datetime64[ms]').astype(int)
    # Convert UTC time array to Unix time array
    unix_millis_array = np.array(utc_millis_array) - unix_epoch_start
    return unix_millis_array
def unix_to_utc_millis_array(unix_millis_array):
    # Unix epoch start time in milliseconds
    unix_epoch_start = np.datetime64('1970-01-01T00:00:00', 'ms').astype('datetime64[ms]').astype(int)
    # Convert Unix time array to UTC time array
    utc_millis_array = np.array(unix_millis_array) + unix_epoch_start
    return utc_millis_array
```

```python
folder_path="C:/Users/ecems/OneDrive/Masaüstü/AI/term_project/test/"
length=0
counter=0
length_array=[]
for i in range(len(sample_sub_['tripId'])):
        phone_path=os.path.join(folder_path+sample_sub_['tripId'][i])
        gnss_file_path = os.path.join(phone_path, "device_gnss.csv")

        selected_columns = ['SvVelocityYEcefMetersPerSecond', 'SvVelocityZEcefMetersPerSecond','SvVelocityXEcefMetersPerSecond', 'RawPseudorangeMeters', 
                            'SvClockBiasMeters', 'IsrbMeters', 'TroposphericDelayMeters','IonosphericDelayMeters','WlsPositionXEcefMeters', 
                            'WlsPositionYEcefMeters', 'WlsPositionZEcefMeters']
        df_gnss=pd.read_csv(gnss_file_path, low_memory=False)
        gnss_reduced = df_gnss.groupby('utcTimeMillis')[selected_columns].apply(lambda x: x.apply(np.random.choice)).reset_index()
        lon1, lat1, alt1 = gnss_reduced.reset_index()['WlsPositionXEcefMeters'],gnss_reduced.reset_index()['WlsPositionYEcefMeters'],gnss_reduced.reset_index()['WlsPositionZEcefMeters']
        
        common_time_array = sample_sub_['UnixTimeMillis'][i]
        common_time_array=unix_to_utc_millis_array(common_time_array)
        
        time=common_time_array//1000
        time_2=np.array(gnss_reduced['utcTimeMillis']//1000)

        df_imu=pd.read_csv(phone_path+"/device_imu.csv")
        df_acc=df_imu.loc[df_imu['MessageType'] == 'UncalAccel']
        df_gyr=df_imu.loc[df_imu['MessageType'] == 'UncalGyro']
        df_acc = df_acc.copy()
        df_gyr = df_gyr.copy()
        df_acc.loc[:, 'utcTimeSeconds'] = df_acc['utcTimeMillis'] // 1000
        df_gyr.loc[:, 'utcTimeSeconds'] = df_gyr['utcTimeMillis'] // 1000
        selected_imu=['MeasurementX','MeasurementY','MeasurementZ']
        summed_values_acc = df_acc[df_acc['utcTimeSeconds'].isin(time)].groupby('utcTimeSeconds')[selected_imu].sum().reset_index()
        summed_values_gyr = df_gyr[df_gyr['utcTimeSeconds'].isin(time)].groupby('utcTimeSeconds')[selected_imu].sum().reset_index()
        # summed_values_acc DataFrame'i için sadece belirli sütunları seçme ve sütun isimlerini değiştirme
        summed_values_acc = summed_values_acc[['MeasurementX', 'MeasurementY', 'MeasurementZ','utcTimeSeconds']]
        summed_values_acc.columns = ['acc_MeasurementX', 'acc_MeasurementY', 'acc_MeasurementZ','utcTimeSeconds']

        # summed_values_gyr DataFrame'i için sadece belirli sütunları seçme ve sütun isimlerini değiştirme
        summed_values_gyr = summed_values_gyr[['MeasurementX', 'MeasurementY', 'MeasurementZ','utcTimeSeconds']]
        summed_values_gyr.columns = ['gyr_MeasurementX', 'gyr_MeasurementY', 'gyr_MeasurementZ','utcTimeSeconds']

        result_gnss = pd.DataFrame({'tripId': sample_sub_['tripId'][i],
                                   'utcTimeMillis': np.array(gnss_reduced['utcTimeMillis']),
                                   'SvVelocityYEcefMetersPerSecond': np.array(gnss_reduced['SvVelocityYEcefMetersPerSecond']),
                                   'SvVelocityZEcefMetersPerSecond': np.array(gnss_reduced['SvVelocityZEcefMetersPerSecond']),
                                   'SvVelocityXEcefMetersPerSecond': np.array(gnss_reduced['SvVelocityXEcefMetersPerSecond']),
                                   'RawPseudorangeMeters': np.array(gnss_reduced['RawPseudorangeMeters']),
                                   'SvClockBiasMeters': np.array(gnss_reduced['SvClockBiasMeters']),
                                   'IsrbMeters': np.array(gnss_reduced['IsrbMeters']),
                                   'TroposphericDelayMeters': np.array(gnss_reduced['TroposphericDelayMeters']),
                                   'IonosphericDelayMeters': np.array(gnss_reduced['IonosphericDelayMeters']),
                                   'WlsPositionXEcefMeters': np.array(gnss_reduced['WlsPositionXEcefMeters']),
                                   'WlsPositionYEcefMeters': np.array(gnss_reduced['WlsPositionYEcefMeters']),
                                   'WlsPositionZEcefMeters': np.array(gnss_reduced['WlsPositionZEcefMeters'])
                                   })

        merged_imu = pd.merge(summed_values_acc, summed_values_gyr, on='utcTimeSeconds', how='left')
        merged_imu = merged_imu.fillna(method='ffill').fillna(method='bfill')
        sub_time_df=pd.DataFrame({'utcTimeMillis': common_time_array})
        result_gnss=result_gnss[result_gnss['utcTimeMillis'].isin(sub_time_df['utcTimeMillis'])]
        merge_sub_gnss=pd.merge(sub_time_df,result_gnss,on='utcTimeMillis', how='left')
        merge_sub_gnss['utcTimeSeconds']=time
        merged_imu=merged_imu[merged_imu['utcTimeSeconds'].isin(time)]
        merged_last=pd.merge(merge_sub_gnss,merged_imu,on='utcTimeSeconds', how='left')
        merged_last=merged_last.fillna(method='ffill').fillna(method='bfill')
        length+=len(merged_last['utcTimeMillis'])
        counter+=1
        length_array.append(len(merged_last['utcTimeMillis']))
        file_name = f'test_data/{counter}.csv'
            
        file_exists = os.path.exists(file_name)
            
        if file_exists:
                file_name = file_name[:-4] + "_1.csv"
                
            # Save the DataFrame to CSV
        merged_last.to_csv(file_name, index=False)
        print("file number : {} total length : {} ".format(counter,length))
        
        if len(time)!=len(merged_last['utcTimeMillis']):
                print("HAYYYAAMMMM")
```
